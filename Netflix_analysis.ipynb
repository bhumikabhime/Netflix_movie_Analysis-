# **Which individual country has the highest number of TV Shows?**

```python
# prompt: which individual country has the highest number of TV Shows

# Find the country with the highest number of TV shows
country_tv_shows = df[df['type'] == 'TV Show']['country'].value_counts().head(1)
highest_country = country_tv_shows.index[0]
print(f"The country with the highest number of TV shows is: {highest_country}")
```
